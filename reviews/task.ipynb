```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("employee_data.csv")
print(df.columns)
ds = df.groupby("Department")[["Salary","Performance_Score"]].mean()
print(ds)
```

```python
# df3 = pd.DataFrame(np.where(df["Salary"] < 50000, "not good salary",np.where(df["Salary"] > 60000 , "very good salary", "good salary")))
# print(type(df3))
# print(df3.value_counts())

def sal_category(salary):
    if salary < 50000:
        return "not good salary"
    elif  salary > 50000 and salary < 60000:
        return "good salary"
    elif salary > 60000:
        return "very good salary"

df2 = df["Salary"].apply(sal_category).value_counts()
print(df2)
```

```python
# (a - min) / (max - min)
#salary
sal = np.array(df["Salary"])
min_sal = sal.min()
max_sal = sal.max()
print(min_sal)
print(max_sal)
normalized_salary = (sal - min_sal) / (max_sal- min_sal)
print(normalized_salary)
#Performance_Score
perf = np.array(df["Performance_Score"])
min_per = perf.min()
max_per = perf.max()
normalized_performance = (perf - min_per) / (max_per - min_per)
print(normalized_performance)
#Experience_years
exp = np.array(df["Experience_Years"])    
min_exp = exp.min()
max_exp = exp.max()
normalized_experience = (exp - min_exp) / (max_exp - min_exp)
print(normalized_experience)
```

```python
# Analyze correlation between salary and performance using numpy
correlation = np.corrcoef(df["Salary"], df["Performance_Score"])
print(correlation)
# print(df[["Salary","Performance_Score"]].corr())
```
