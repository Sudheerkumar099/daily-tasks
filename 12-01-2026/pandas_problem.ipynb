```python
# Pandas (Intermediate)
# A DataFrame contains date, city, temperature, and humidity.
# Task:
# Filter data for a single city
# Compute rolling 3-day average temperature
# Detect days where humidity increased compared to the previous day
import pandas as pd
data = {
    "date": [
        "2025-01-01", "2025-01-01", "2025-01-01",
        "2025-01-02", "2025-01-02", "2025-01-02",
        "2025-01-03", "2025-01-03", "2025-01-03"
    ],
    "city": [
        "Bengaluru", "Hyderabad", "Chennai",
        "Bengaluru", "Hyderabad", "Chennai",
        "Bengaluru", "Hyderabad", "Chennai"
    ],
    "temperature": [24, 26, 28, 23, 27, 29, 22, 25, 30],
    "humidity": [68, 60, 72, 70, 58, 75, 65, 62, 78]
}

df = pd.DataFrame(data)
blr = df[df["city"] == "Bengaluru"]
hyd = df[df["city"] == "Hyderabad"]
chn = df[df["city"] == "Chennai"]
print("Filtered Data of Bengalulu")
print(blr)
print("Filtered Data of Hyderabad")
print(hyd)
print("Filtered Data of chennai")
print(chn)

# average temperature 
print(blr["temperature"].mean())
print(chn["temperature"].mean())
print(hyd["temperature"].mean())
```

```python
for i in df.index:
   previous = df.loc[i,"humidity"]
   rows = df.shape
   if i < rows[0]-1:
      current = df.loc[i+1,"humidity"]
      print(previous,current)
      if current > previous :
         print(df.loc[i+1,"date"])
```
