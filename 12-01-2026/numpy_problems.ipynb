```python
import numpy as np
# NumPy (Intermediate)
#  A weather station stores hourly temperature readings for 7 days in a NumPy array of shape (7, 24).
# Task:
# Find the daily maximum and minimum temperature
# Identify the day with the largest temperature variation
# Replace outliers beyond Â±2 standard deviations with the mean temperature
temperatures = np.random.randint(1,100,(7,24))
print(temperatures)
highest = 0
day = 0
for i in range(0,7):
    high_temp = temperatures[i,0::].max()
    min_temp = temperatures[i,0::].min()
    print(f"Temperatures of day {i+1} is \n max temperature = {high_temp} \n min temperature = {min_temp}")
    if high_temp >= highest:
        highest = high_temp
        day = i+1
print(f"Day: {day} is the day with highest Temperature with {highest} Among 7 days")
mean = temperatures.mean()
print(mean)
std = temperatures.std()
sd1 =  2 * std
sd2 =  -2 * std
temperatures = np.array(np.where(temperatures > sd1, mean, temperatures),dtype = int)
temperatures = np.array( np.where(temperatures < sd2, mean, temperatures),dtype = int)
```
